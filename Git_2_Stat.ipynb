```python
# Однофакторным дисперсионный анализ

# QQ plot
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt    #from pylab import *  

# Generate some data for this demonstration.
data = stats.norm.rvs(loc=10, scale=2.5, size=30)

# Fit a normal distribution to the data:
mu, std = stats.norm.fit(data)

fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 9))

# Q-Q plot (probability plot)
stats.probplot(data, plot=ax1)

# Plot the histogram.
ax2.hist(data, bins=25, density=True, alpha=0.6, color='g')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
ax2.plot(x, p, 'k', linewidth=2)

print(xmin , xmax, x[1:5], x[-5:] , p[1:5], p[-5:-1] )
''' 
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
'''
```

```python
# Дальше то же самое , но посчитано вручную
mean = data.expr.mean()
print("mean", mean)

SST = 0
for i in data.expr:
  step = pow(i - mean, 2)
  SST += step
print("SST:", SST) 


SSW = 0.0
SSB = 0.0
group_mean = 0.0
groups = data.Therapy.drop_duplicates()

for group_name in groups: 
    group = data[data.Therapy==group_name]
    group_mean = group.expr.mean()
    SSB += group.expr.count() * pow(group_mean - mean, 2)
    for i in group.expr:
        SSW += pow(i - group_mean, 2)
print("SSW:", SSW)
print("SSB:", SSB)

df_within = data.expr.count() - groups.count()
df_beetwen = groups.count() - 1
F = (SSB / df_beetwen) / (SSW / df_within)
print("F:", F)


from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(20,8))
dfn, dfd = df_beetwen,df_within
x = np.linspace(0, F+1)
plt.plot(x, f.pdf(x, dfn, dfd),'r-', lw=5, alpha=0.6, label='f pdf')
plt.axvline(x=F, label='F-score')
plt.legend()

print("P:", f.pdf(F, dfn, dfd))
```
